```python
import numpy as np
import pandas as pd
```

```python
DATA=pd.read_csv('data.csv')
data=DATA.copy()
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()      #gives the total no. of null value
```

```python
X = data.iloc[:, 1:43].values       # Gives values until the last column
y = data.loc[:, 'score'].values         # Gives only the last column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 

# here the "test_size=0.2" indicates that 20% of the data will be kept as the test set and the remaining as training set.
```

```python
from sklearn.linear_model import LinearRegression  

regressor = LinearRegression()            # This function is assigned to the variable 'regressor'
regressor.fit(X_train, y_train)           # This function is fitted with x-train and y_train on which the model will be trained

print("Training complete")
```

```python
y_pred = regressor.predict(X_test) 
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

# MODEL ACCURACY: 99.42%
